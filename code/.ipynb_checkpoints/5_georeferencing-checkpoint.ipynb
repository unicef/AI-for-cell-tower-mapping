# Georeferencing
This script outputs two geojson files:
*   `geo_boxes_{MODEL}_{SPLIT}.geojson` file contains bounding with lat/lng
*   `centroid_{MODEL}_{SPLIT}.geojson` file contains centroids of bboxes in lat/long



```python
import geopandas as gpd
from shapely.geometry import box
from shapely.affinity import affine_transform
from rasterio.transform import from_bounds
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import geopy.distance

import json
import numpy as np
import os
```

```python
# Mount google drive connection to have access to the shared data
from google.colab import drive
drive.mount('/content/drive')
```

```python
def return_file_name(img_id, image_map):
  """
  Get image file name using image id
  Args:
    image_map (list): List of dictionaries, each dictionary containing
    id (img id) and file_name
  Returns
    string: returns the name of the image file
  """
  img_files = [img['file_name'] for img in image_map if img['id'] == img_id]
  if len(img_files) > 1:
    raise ValueError('Image id appears in more than one file')
  elif len(img_files) == 0:
    raise ValueError("Image doesn't exits")
  else:
    return img_files[0]

  
def from_imgcoords_to_latlong(
  bbox_coords, img_lat_long, 
  img_measurements=[512,512]
  ):
  """
  Args:
    img_coords (list): list with [x_min, y_min, x_max, y_max]
    img_lat_long (list): lat long coordinates of the image with 
    the format [top, left, bottom, rigth]
    img_measurements (list): [img_width, img_height]
  Returns:
    Returns a transformed geometry using an affine transformation matrix
  """
  bbox = box(bbox_coords[0], bbox_coords[1], bbox_coords[2], bbox_coords[3])
  affine_obj = from_bounds(
      north = img_lat_long[0],
      west = img_lat_long[1], 
      south = img_lat_long[2], 
      east = img_lat_long[3], 
      width = img_measurements[0], 
      height = img_measurements[1]
      )

  geom_bbox = affine_transform(
      bbox, 
      [affine_obj.a, affine_obj.b,
       affine_obj.d, affine_obj.e,
       affine_obj.xoff, affine_obj.yoff]
       )
  return geom_bbox
```

```python
# Read files and define paths
############### Fill out this section ###############
MODEL = 'yolo'
SPLIT = 'val'
#####################################################


metadata_path = '/content/drive/MyDrive/UNICEF & NYU Giga initiative - data sharing/data/metadata_splits.xlsx'
results_path = '/content/drive/MyDrive/UNICEF & NYU Giga initiative - data sharing/data/results'
resized_image_path = '/content/drive/MyDrive/UNICEF & NYU Giga initiative - data sharing/data/interim_images/resized_images_512'
bbox_path = '/content/drive/MyDrive/UNICEF & NYU Giga initiative - data sharing/data/bounding_boxes'
img_latlong_path = '/content/drive/MyDrive/UNICEF & NYU Giga initiative - data sharing/Satellite images/3k_with_bboxes_corrected_nodeIds.xlsx'
COCO_ANNO_PATH = os.path.join(results_path, f'{MODEL}_{SPLIT}.json')
COCO_TARGET_PATH = os.path.join(bbox_path, f'{SPLIT}.json')

metadata = pd.read_excel(metadata_path)
img_latlong_data = pd.read_excel(img_latlong_path)
bbox_pred = json.load(open(COCO_ANNO_PATH))
bbox_target = json.load(open(COCO_TARGET_PATH))
image_map = bbox_target['images']
```

```python
geom_list_geo = []
node_ids = []
for bbox in bbox_pred:
  # Getting node id
  bbox_img = return_file_name(
      bbox['image_id'],
      image_map
      )
  node_id = int(bbox_img.replace('.png', ''))
  
  # Getting lat long of the image
  img_latlong = img_latlong_data.loc[
    img_latlong_data['OSM Node ID'] == node_id, 
    ['top', 'left', 'bottom', 'right']
    ].to_numpy()
  if img_latlong.shape[0] == 0:
    raise ValueError("Node id doesn't exist")
  if img_latlong.shape[0] > 1:
    raise ValueError('Node id tied to more than one image')
  img_latlong = img_latlong[0]

  # Converting bbox from coco to xmin, ymin, xmax, ymax
  bbox_coco = bbox['bbox']
  bbox_kitti = [
      bbox_coco[0],
      bbox_coco[1],
      bbox_coco[0] + bbox_coco[2],
      bbox_coco[1] + bbox_coco[3]
      ]

  # Transforming bbox into lat/long coords
  geom_bbox = from_imgcoords_to_latlong(
        bbox_kitti,
        img_latlong
    )
  geom_list_geo.append(geom_bbox)
  node_ids.append(node_id)
```

```python
crs = "EPSG:4326"
geo_df = gpd.GeoDataFrame(
    node_ids, 
    columns=['node_id'],
    geometry=geom_list_geo,
    crs=crs
    )
# save geojson file
geo_df.to_file(
    os.path.join(results_path, f'geo_boxes_{MODEL}_{SPLIT}.geojson')
    )

centroid_df = gpd.GeoDataFrame(
    node_ids, 
    columns=['node_id'],
    # Projecting into a flat space before calcularing centroid (see issue here https://gis.stackexchange.com/questions/372564/userwarning-when-trying-to-get-centroid-from-a-polygon-geopandas)
    geometry=geo_df.to_crs('+proj=cea').centroid.to_crs(geo_df.crs),
    crs=crs
    )
centroid_df.to_file(
    os.path.join(results_path, f'centroids_{MODEL}_{SPLIT}.geojson')
    )
```

```python
target_centroids = metadata.loc[
    metadata['Train/Test/Val'] == SPLIT
    ].filter(['OSM Node ID', 'Latitude', 'Longitude']
             ).rename(columns={'OSM Node ID':'node_id'})
```

```python
comparison = target_centroids.merge(
    centroid_df,
    on='node_id',
    how='left'
    )
```

```python
comparison = comparison.assign(
    predicted_lon=comparison.geometry.map(
        lambda y: y.x if y else np.nan),
    predicted_lat=comparison.geometry.map(
        lambda y: y.y if y else np.nan)
        )
```

```python
def distance_between_points(target, predicted):
  if (not np.isnan(predicted[0])) and (not np.isnan(predicted[1])):
    return geopy.distance.geodesic(target, predicted).m
  else:
    return np.nan
```

```python
comparison = comparison.assign(
    distance=comparison.apply(
        lambda x: distance_between_points(
            (x['Latitude'], x['Longitude']),
            (x['predicted_lat'], x['predicted_lon'])),axis=1)
    )
```

```python
comparison.distance.describe()
```

```python
comparison.distance.hist(bins=30)
plt.title('Distance in meters between target and prediction')
```

```python
comparison.groupby('node_id').predicted_lon.count().hist()
plt.title('Num of bounding boxes per image')
```
